```python
wcnf = WCNF()

graph = nx.DiGraph()
graph.add_edge(1, 2)
graph.add_edge(2, 3)
graph.add_edge(3, 4)
graph.add_edge(4, 1)

cycles = list(nx.simple_cycles(graph))
n = graph.number_of_nodes()
#print(cycles)

wcnf = WCNF()
#pairs = []
m = 0

for cycle in cycles:
    arr = list(map(int, cycle))
    add = []
    for i in range(len(arr)):
        x, y = arr[i], arr[(i+1)%len(arr)]
        #print(x,y)
        index = n*x + y
        #print(index)
        add.append(index)
        m = max(m, index)
        #pairs.append((x, y))
    wcnf.append(add)

for i in range(1, m+1):
    wcnf.append([-i], weight=1)

print(m)

#for pair in pairs:
#    wcnf.append([-(n*pair[0]+pair[1])], weight=1)

sol_g = RC2(wcnf)
#print(sol_g.compute())
opt = next(sol_g.enumerate())
print(opt)

remove = []
for i in range(1, n+1):
    if i in opt:
        remove.append(i)
print(remove)

'''wcnf.append([1, 2, 3, 4])
wcnf.append([1, -3])
wcnf.append([-1], weight=1)
wcnf.append([-2], weight=1)
wcnf.append([-3], weight=1)
wcnf.append([-4], weight=1)'''

with RC2(wcnf) as rc2:
    i = 0
    for m in rc2.enumerate():
        print(m)
        for edge in m:
            if edge > 0:
                y = n if (edge) % n == 0 else (edge) % n
                x = (edge - y) // n
                print('removing edge {0} {1}'.format(x, y))
        #print('model {0} has cost {1}'.format(m, rc2.cost))
        if i >= 5:
            break
        i += 1
```
