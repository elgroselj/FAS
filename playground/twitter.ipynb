```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.DiGraph()
with open("twitter_combined_sample.txt", "r") as f:
    print("INFO: Reading and adding edges from file ...")
    while True:
        line = f.readline()
        if len(line) == 0:
            break
        v1, v2 = map(int, line.split(" "))
        graph.add_edge(v1,v2)
```

```python
nx.write_pajek(graph, "twitter_combined_sample.net")
```

```python
# nodeId.edges : [...] The 'ego' node does not appear, but it is assumed that they follow every node id that appears in this file.
# adding ego node edges in not needed, as ego node is not part of any strongly connected component anyway


graph = nx.DiGraph()
with open("twitter_combined.txt", "r") as f:
    print("INFO: Reading and adding edges from file ...")
    while True:            
        line = f.readline()
        if len(line) == 0:
            break
        v1, v2 = map(int, line.split(" "))
        graph.add_edge(v1,v2)
    print("INFO: Done reading and adding edges from file.")
        

# nx.draw(graph,node_size=10)
```

```python
print("Number of edges: ", len(list(graph.edges())))
```

```python
# for scc in nx.strongly_connected_components(graph):
#     if len(scc) > 1:
        # print(scc)
        # print(len(scc))
        
plt.hist([len(scc) for scc in nx.strongly_connected_components(graph)],bins=100)
plt.yscale("log")
plt.show()
# with filter: if len(scc) < 60000
plt.hist([len(scc) for scc in nx.strongly_connected_components(graph) if len(scc) < 60000],bins=100)
plt.yscale("log")
plt.show()
```

```python
nx.write_pajek(graph, "twitter_combined.net")
```

```python
H_nodes = [scc for scc in nx.strongly_connected_components(graph) if len(scc) > 60000][0]
H = graph.subgraph(H_nodes)
len(list(H.nodes()))
```

```python
nx.write_pajek(H, "twitter_combined_largest_scc.net")
```
