# Diff between LB and UB

```python
df = pd.read_csv("../nothuge_tw_iqr_gap_UB_diversesfill_diversesfill_smartAEainf_diverses_nonodes_sols_4_3_2_1_components.csv")
```

```python
df["interval"] = df["UB"] - df["LB"]
df["min"] = df.apply(lambda x: min( x["remove_edges_FAS_greedy_sol"],
                                    x["remove_edges_pagerank_scc_iteratively_sol"],
                                    x["remove_edges_trueskill_scc_iteratively_sol"],
                                    x["diverses_sol"],
                                    x["smartAE_ainf_sol"]), axis=1)
df["max"] = df.apply(lambda x: max( x["remove_edges_FAS_greedy_sol"],
                                    x["remove_edges_pagerank_scc_iteratively_sol"],
                                    x["remove_edges_trueskill_scc_iteratively_sol"],
                                    x["diverses_sol"],
                                    x["smartAE_ainf_sol"]), axis=1)

df['ratio'] = df["max"]/df["min"]
```

```python
# plt.yscale("log")
plt.hist(df["interval"],bins = 100)
```

```python
df = df[(df["n"] <= 5000) & (df["m"] <= 10000)]
```

```python
# df["interval_class"] = "unknown" * len(df)
# df.loc[df.interval==0, 'interval_class'] = "optimal"
# df.loc[(df["interval"]>0) & (df.interval<=10), 'interval_class'] = "near optimal"
# # df[df["interval"] == 0] = "optimal"
# # df[(df["interval"] > 0) &(df["interval"] <= 10)] = "near optimal"Å¾

# Initialize the 'interval_class' column with 'unknown'
# df["interval_class"] = "unknown"
def f(x):
    if x == 0:
        return "= 0 (optimal)"
    if x <= 10:
        return "<= 10"
    if x <= 100:
        return "<= 100"
    return "> 100"
    
# # Assign 'optimal' to rows where 'interval' is 0
# df.loc[df["interval"] == 0, 'interval_class'] = "optimal"

# # Assign 'near optimal' to rows where 'interval' is greater than 0 and less than or equal to 10
# df.loc[(df["interval"] > 0) & (df["interval"] <= 10), 'interval_class'] = "near optimal"

# # Display the DataFrame to check the results
# print(df)

df["interval_class"] = df["interval"].apply(f)
```

```python

category_counts = df['interval_class'].value_counts()

# Apply a Seaborn theme
sns.set_theme()

# Create the pie chart
plt.figure(figsize=(8, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('viridis', len(category_counts)), textprops={'fontsize': 18})

# Add a title
# plt.title('Gap between LB and UB')

# Show the plot
plt.savefig("../images/LB_UB_pie.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

def heuristic_to_col(heuristic):
    if heuristic == "diverses":
        return "yellowgreen"
    elif heuristic == 'remove_edges_FAS_greedy':
        return 'gold'
    elif heuristic == 'remove_edges_trueskill_scc_iteratively':
        return 'lightskyblue'
    elif heuristic == 'remove_edges_pagerank_scc_iteratively':
        return 'lightcoral'
    elif heuristic == 'smartAE_ainf':
        return 'red'
    else:
        return 'black'

figure, axis = plt.subplots(3, 3) 

i = 0
for sub_df in data:
    x = i // 3
    y = i % 3
    category_counts_ = sub_df[1]['best_solver'].value_counts()
    #print(category_counts)

    # Apply a Seaborn theme
    sns.set_theme()
    patches_, texts = axis[x,y].pie(category_counts_, colors=list(map(heuristic_to_col, list(category_counts_.index))), startangle=90)
    axis[x,y].axis('equal')
    axis[x,y].set_title(sub_df[0])


    #plt.savefig("../images/best_pie_" + sub_df[0] + ".png")
    
    i += 1
axis[2,2].axis('off')
axis[2,1].axis('off')
figure.legend(patches, map( lambda x : f"{x[0]}: {x[1]}", zip(category_counts.index,category_counts.values.tolist())), loc="lower right")
plt.tight_layout()
plt.savefig("../images/best_pie_categories.png")
plt.show()
```
